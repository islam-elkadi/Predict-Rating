```python
import re
from collections import Counter
```

```python
with open("./Dataset/training_processed/training_balanced.txt", "r") as f:
    text = f.read()
    text = re.sub(r"__label__\d","", text)
    text = text.split()
    text = Counter(text)
```

```python
text = sorted(text.items())
```

```python
print(text[:][2])
```

```python
for value in text:
    if value[1] > 10000 and not 10000 < value[1] < 35000 and len(value[0]) > 2:
        print(value[0], "{:,}".format(value[1]))
```
